# Customer Churn Prediction - Exploratory Data Analysis (EDA)
This notebook contains the exploratory analysis of the Telco Customer Churn dataset. We will analyze target distribution, missing values, correlation, and analyze how features like tenure, contract type, payment method, and monthly charges affect customer churn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 12
```

## 1. Load the Preprocessed Dataset

```python
df = pd.read_csv('../data/cleaned_telco.csv')
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Basic Dataset Info & Missing Values

```python
print("--- Dataset Info ---")
print(df.info())

print("\n--- Descriptive Statistics ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())
```

## 3. Target Variable Analysis: Churn Distribution

```python
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100
print("Churn counts:")
print(churn_counts)
print("\nChurn percentage:")
print(churn_pct)

plt.figure(figsize=(6, 5))
sns.countplot(x='Churn', data=df, palette='viridis')
plt.title('Distribution of Customer Churn')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks([0, 1], ['Retained (No)', 'Churned (Yes)'])
plt.tight_layout()
plt.savefig('../reports/churn_distribution.png')
plt.show()
```

## 4. Key Numeric Features: Tenure & Monthly Charges

```python
# Monthly Charges distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, color='skyblue')
plt.title('Monthly Charges Distribution')
plt.xlabel('Monthly Charges ($)')

# Tenure distribution
plt.subplot(1, 2, 2)
sns.histplot(df['tenure'], bins=30, kde=True, color='salmon')
plt.title('Tenure Distribution')
plt.xlabel('Tenure (Months)')

plt.tight_layout()
plt.savefig('../reports/numeric_distributions.png')
plt.show()
```

## 5. Contract Type vs. Customer Churn

```python
contract_churn = df.groupby('Contract')['Churn'].mean().reset_index()
contract_churn['Churn'] = contract_churn['Churn'] * 100
print(contract_churn)

plt.figure(figsize=(8, 5))
sns.barplot(x='Contract', y='Churn', data=contract_churn, palette='coolwarm')
plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Churn Rate (%)')
plt.tight_layout()
plt.savefig('../reports/contract_vs_churn.png')
plt.show()
```

## 6. Payment Method vs. Customer Churn

```python
pm_churn = df.groupby('PaymentMethod')['Churn'].mean().reset_index()
pm_churn['Churn'] = pm_churn['Churn'] * 100
pm_churn = pm_churn.sort_values(by='Churn', ascending=False)
print(pm_churn)

plt.figure(figsize=(10, 5))
sns.barplot(x='Churn', y='PaymentMethod', data=pm_churn, palette='magma')
plt.title('Churn Rate by Payment Method')
plt.xlabel('Churn Rate (%)')
plt.ylabel('Payment Method')
plt.tight_layout()
plt.savefig('../reports/payment_vs_churn.png')
plt.show()
```

## 7. Numeric Feature Correlation Heatmap

```python
# Calculate correlation of numeric features
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.savefig('../reports/correlation_matrix.png')
plt.show()
```
